```python
def create_installation_doc(product: str, platform: str) -> Dict:
    """Create installation guide"""
    return {
        'id': f"install_{product.lower().replace(' ', '_')}_{platform.lower()}",
        'title': f"How to Install {product} on {platform}",
        'product': product,
        'category': 'installation',
        'platform': platform,
        'content': f"""# Installing {product} on {platform}

## System Requirements
- {platform} operating system (latest version recommended)
- At least 2GB of free disk space
- Internet connection for download and activation

## Installation Steps

1. **Download the installer**
   - Visit our official website at www.example.com
   - Navigate to Downloads section
   - Select {product} for {platform}

2. **Run the installer**
   - Locate the downloaded file in your Downloads folder
   - Double-click to run the installer
   - Follow the on-screen instructions

3. **Complete setup**
   - Enter your license key when prompted
   - Choose installation directory (default recommended)
   - Click 'Install' and wait for completion

4. **First launch**
   - Launch {product} from your applications menu
   - Sign in with your account credentials
   - Complete the initial setup wizard

## Troubleshooting Installation Issues

If you encounter any problems during installation:
- Ensure you have administrator privileges
- Disable antivirus temporarily
- Check firewall settings
- Contact support if issues persist
""",
        'tags': ['installation', 'setup', 'getting-started'],
        'created_at': datetime.now().isoformat()
    }

def create_troubleshooting_doc(product: str, issue: str, solution: str) -> Dict:
    """Create troubleshooting guide"""
    doc_id = f"trouble_{product.lower().replace(' ', '_')}_{issue.lower().replace(' ', '_')}"
    return {
        'id': doc_id,
        'title': f"{product}: {issue}",
        'product': product,
        'category': 'troubleshooting',
        'content': f"""# {issue} in {product}

## Problem Description
{issue}

## Solution
{solution}

## Additional Tips
- Make sure you're running the latest version of {product}
- Check system requirements are met
- Try restarting the application
- Contact support if the issue persists
""",
        'tags': ['troubleshooting', 'problem', 'fix'],
        'created_at': datetime.now().isoformat()
    }

def create_feature_doc(product: str, feature: str, description: str) -> Dict:
    """Create feature documentation"""
    doc_id = f"feature_{product.lower().replace(' ', '_')}_{feature.lower().replace(' ', '_')}"
    return {
        'id': doc_id,
        'title': f"{product}: {feature}",
        'product': product,
        'category': 'features',
        'content': f"""# {feature} in {product}

## Overview
{description}

## How to Use
1. Open {product}
2. Navigate to the {feature} section
3. Follow the in-app instructions

## Tips and Best Practices
- Use this feature regularly to get the most value
- Explore all available options in settings
- Check our video tutorials for detailed walkthroughs
""",
        'tags': ['features', 'how-to', 'guide'],
        'created_at': datetime.now().isoformat()
    }

def create_faq_doc(product: str, question: str, answer: str) -> Dict:
    """Create FAQ Ð·Ð°Ð¿Ð¸Ñ"""
    doc_id = f"faq_{product.lower().replace(' ', '_')}_{question[:20].lower().replace(' ', '_')}"
    return {
        'id': doc_id,
        'title': f"FAQ: {question}",
        'product': product,
        'category': 'faq',
        'content': f"""# Frequently Asked Question

## Question
{question}

## Answer
{answer}
""",
        'tags': ['faq', 'question', 'answer'],
        'created_at': datetime.now().isoformat()
    }

print("Document templates created")
```

```python
os.makedirs('data/raw', exist_ok=True)
os.makedirs('data/processed', exist_ok=True)

json_path = 'data/raw/knowledge_base.json'
with open(json_path, 'w', encoding='utf-8') as f:
    json.dump(documents, f, indent=2, ensure_ascii=False)

print(f"Saved {len(documents)} documents to {json_path}")

docs_dir = 'data/raw/documents'
os.makedirs(docs_dir, exist_ok=True)

for doc in documents:
    filename = f"{doc['id']}.md"
    filepath = os.path.join(docs_dir, filename)
    
    with open(filepath, 'w', encoding='utf-8') as f:
        f.write(doc['content'])

print(f"Saved {len(documents)} individual markdown files to {docs_dir}/")

metadata = [{
    'id': doc['id'],
    'title': doc['title'],
    'product': doc['product'],
    'category': doc['category'],
    'tags': doc['tags'],
    'created_at': doc['created_at']
} for doc in documents]

metadata_path = 'data/raw/metadata.json'
with open(metadata_path, 'w', encoding='utf-8') as f:
    json.dump(metadata, f, indent=2, ensure_ascii=False)

print(f"Saved metadata to {metadata_path}")

print("\n" + "="*80)
print("Data generation complete!")
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

products = list(product_counts.keys())
counts = list(product_counts.values())
ax1.bar(products, counts, color=['#3498db', '#2ecc71', '#f39c12', '#e74c3c'])
ax1.set_xlabel('Product')
ax1.set_ylabel('Number of Documents')
ax1.set_title('Documents by Product')
ax1.tick_params(axis='x', rotation=45)

categories = list(category_counts.keys())
cat_counts = list(category_counts.values())
ax2.bar(categories, cat_counts, color='#9b59b6')
ax2.set_xlabel('Category')
ax2.set_ylabel('Number of Documents')
ax2.set_title('Documents by Category')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('data/raw/data_statistics.png', dpi=150, bbox_inches='tight')
plt.show()

print("Statistics chart saved to data/raw/data_statistics.png")
```
